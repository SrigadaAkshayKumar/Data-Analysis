# DATA VISUALIZATION

## MATPLOTLIB

```python
import matplotlib.pyplot as plt
```

## Ploting Types in Matplotlib

# 1. Line Chart

## Sales over state

```python
sales_over_state = df1.groupby('State')['Sales'].sum()
plt.figure(figsize=(12,5))
plt.plot(sales_over_state.index, sales_over_state.values, marker='*', linestyle='-', color='g')
plt.title('Sales Over States')
plt.xlabel('State Name')
plt.ylabel('Total Sales')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

# 2. Bar Chart

## Region wise sales

```python
region_wise_sales=df1.groupby('Region')['Sales'].sum()
plt.figure(figsize=(8,5))
plt.bar(region_wise_sales.index, region_wise_sales.values, color="orange")
plt.title('Region Wise Sales')
plt.xlabel('Region')
plt.ylabel('sales')
plt.show()
```

# 3. scatter plot

## Sales vs Profit

```python
plt.figure(figsize=(7,5))
plt.scatter(df1['Sales'], df1['Profit'], color='green')
plt.title('Sales vs Profit')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.show()
```

# 4. Histogram

```python
plt.figure(figsize=(8, 5))
count, bins, ignored = plt.hist(df['Profit'], bins=20, color="orange", alpha=0.7, edgecolor='black')
plt.plot(bins[:-1], count, linestyle='-', color='blue')
plt.title('Distribution of Profits')
plt.xlabel('Profit')
plt.ylabel('Frequency')
plt.show()
```

# 5. Pie Chart

## Sales Distribution by State

```python
data = df.groupby('State')['Sales'].sum()
plt.figure(figsize=(8, 8))
plt.pie(data, labels=data.index, autopct='%1.1f%%')
plt.title('Sales Distribution by State') 
plt.xticks(rotation=90)
plt.show()
```

# 6. Heat Map

## Seaborn

```python
import seaborn as sns
```

```python
data = df.pivot_table(index='State', columns='Category', values='Sales', aggfunc='sum')
plt.figure(figsize=(10, 8))
sns.heatmap(data, cmap='viridis', annot=True, fmt='.0f', cbar_kws={'label': 'Total Sales'})
plt.title('Sales by State and Category')
plt.show()
```
