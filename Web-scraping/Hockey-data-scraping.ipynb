```python
from bs4 import BeautifulSoup
import requests
```

```python
url='https://www.scrapethissite.com/pages/forms/'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
soup.find_all('tr')
```

```python
data = soup.find_all('th')
```

```python
data
```

```python
table_titles = [title.text.strip() for title in data]
print(table_titles)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns=table_titles)
```

```python
df
```

```python
column_data = soup.find_all('td')
```

```python
column_data
```

```python
import re
data_list = []

for item in column_data:
    text = item.get_text(strip=True)
    if text == '':
        data_list.append(None)  
    else:
        if re.match(r'^-?\d+(\.\d+)?$', text):
            text = float(text) if '.' in text else int(text)
        data_list.append(text)
print(data_list)
```

```python
import pandas as pd
n_cols = 9
rows = [data_list[i:i+n_cols] for i in range(0, len(data_list), n_cols)]
df = pd.DataFrame(rows, columns=['Name', 'Year', 'Wins', 'Losses', 'OT Losses', 'PCT', 'GF', 'GA', 'Diff'])
df
```
